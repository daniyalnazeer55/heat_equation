```python
# Importing libraries to work with:

import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(7,4)
plt.style.use('ggplot')
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context(context='notebook',
                rc=None)
```

**EXPLICIT METHOD TO SOLVE HEAT EQUATION:**
Solution found through explicit method to finding temperature is:

T(x,t+k) = λT(x-h,t) + (1-2λ)T(x,t) + λT(x+h,t)

$T_{i,j}$ = λ$T_{i-1,j-1}$ + (1-2λ)$T_{i,j-1}$ + λ$T_{i+1,j-1}$  

where λ =  k/h^2

```python
# Initilize your position (x-axis) and time (y-axis positions)
# You can experiment with h and k values to change lambda

h = 0.025
k = 0.025
x = np.arange(0,1+h,h).round(3)
t = np.arange(0,0.1+k,k).round(3)
print('_'*102)
print(f'Position vector is given by: {x}')
print('_'*56)
print(f'Time vector is given by: {t}')

# Initialize your solution matrix

rows = len(x)
cols = len(t)
print('_'*37)
T_matrix = np.zeros((rows,cols))
print(f'Random Initialized Solution Matrix is: \n{T_matrix}')
print('_'*36)
factor = k/h**2
print(f'Value of lambda is: {factor}')
```

```python
# Initialize the boudry conditions:

BoundryConditions = [0,0]
InitialCondtitions = np.sin(np.pi*x)

T_matrix[0,:] = BoundryConditions[0]
T_matrix[-1,:] = BoundryConditions[1]
T_matrix[:,0] = InitialCondtitions

print(f'Updated Matrix with Inital and Boundry Conditions:\n\n\
{T_matrix.round(3).T}')
```

```python
# Caluclating Inner values:

for j in range(1,cols):
  for i in range(1,rows-1):
    T_matrix[i,j] = factor*T_matrix[i-1,j-1]+(1-2*factor)*T_matrix[i,j-1]\
    + factor*T_matrix[i+1,j-1]

print('Solution Matrix:')
print('_'*27)
T_matrix.round(3)
```

```python
R = np.linspace(1,0,cols)
B = np.linspace(0,1,cols)
G=0

for line in range(cols):
  plt.plot(x,T_matrix[:,line],
           color=[R[line],G,B[line]])

plt.title('Plot of Temperature$\degree$C as function of Time(s)',
          fontweight='bold')
plt.xlabel('Distance (m)')
plt.ylabel(r'Temperature $\degree$C')
plt.legend([f't={value}s' for value in t],
           bbox_to_anchor=(1,1.02),loc='upper left')

plt.show()
```
