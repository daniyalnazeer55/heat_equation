```python
# Importing Libraries

import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
# Setting Parameters
L = 3.14 #Length
low_lim = 0 #Lower limit for integration
up_lim = L # Upper limit for integration
```

```python
# Defining integrand and then integrating with respect to given f(x)

def beta(x):
    return (2 / L) * quad(lambda x: f(x) * np.sin((np.pi * x) / L), low_lim, up_lim)[0]

def u(x, t):
    return beta(x) * np.sin((np.pi * x) / L) * np.exp(-(((alpha * np.pi) / L) ** 2) * t)
```

```python
# Choose a certain value of alpha and set your f(X)

alpha = 0.025
def f(x):
    return np.exp(-x)
x = np.linspace(0, 10, 100)
t = np.linspace(0,10 , 100)
X, T = np.meshgrid(x, t)
Z = u(X, T)
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, T, Z, cmap='magma')
plt.title('f(x) = exp(-x)')
ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
# Show the plot
plt.show()
```
