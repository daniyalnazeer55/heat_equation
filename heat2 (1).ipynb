```python
# Importing Libraries

import numpy as np
from sympy import *
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
# Defining constants for integration and calculating U(x,t)

alpha = 0.1
L = 3
up_lim = L
low_lim = 1
v = 3
```

```python
# Breaking U(x,t) to ease calculations:

def f(x):
   return np.cos(x)
def g(x):
  return 1
def factor_1(x):
    return (2 / L) * quad(lambda x: f(x) * np.sin(np.pi * x / L), low_lim, up_lim)[0]
def factor_2(x, f=lambda x: 1):
    return (2 / L) * quad(lambda x: g(x) * np.sin(np.pi * x / L), low_lim, up_lim)[0]
def u(x,t):
  return (((v*L)/(np.pi))*factor_1(x)*np.sin((np.pi*x)/L)*np.sin((np.pi*t)/(v*L)))+\
   (factor_2(x)*np.sin((np.pi*x)/L)*np.cos((np.pi*t)/(v*L)))


# Plotting U(x,t)
plt.rcParams['figure.figsize']=(12,7)
x = np.linspace(1, 20, 30)
t = np.linspace(1, 20, 30)
X, T = np.meshgrid(x, t)
Z = u(x=X,t=T)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, T, Z, 50, cmap='magma')
plt.title('U(x,t) for f(x)=cos(x)', fontweight='bold')
ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
plt.show()
```
